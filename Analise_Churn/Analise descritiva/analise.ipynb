## Dicionário dos dados

- `ID do Cliente`: um ID único que identifica cada cliente.<br>
- `Gênero`: O gênero do cliente: Masculino, Feminino.<br>
- `Idade`: A idade atual do cliente, em anos, no momento em que o trimestre fiscal terminou (2 trimestre 2022).<br>
- `Casado`: Indica se o cliente é casado: Sim, Não.<br>
- `Número de Dependentes`: Indica o número de dependentes que vivem com o cliente (dependentes podem ser filhos, pais, avós, etc).<br>
- `Cidade`: A cidade de residência principal do cliente na Califórnia.<br>
- `Código Postal`: O código postal da residência principal do cliente.<br>
- `Latitude`: A latitude da residência principal do cliente.<br>
- `Longitude`: A longitude da residência principal do cliente.<br>
- `Número de Indicações`: Indica o número de vezes que o cliente indicou um amigo ou membro da família para esta empresa até o momento.<br>
- `Tempo de Permanência em Meses`: Indica a quantidade total de meses que o cliente esteve com a empresa até o final do trimestre especificado acima.<br>
- `Oferta`: Identifica a última oferta de marketing que o cliente aceitou: Nenhuma, Oferta A, Oferta B, Oferta C, Oferta D, Oferta E.<br>
- `Serviço de Telefone`: Indica se o cliente assina o serviço de telefone residencial com a empresa: Sim, Não.<br>
- `Custo Mensal Médio de Chamadas de Longa Distância`: Indica o custo médio de chamadas de longa distância do cliente, calculado até o final do trimestre especificado acima (se o cliente não estiver inscrito no serviço de telefone residencial, isso será 0).<br>
- `Linhas Múltiplas`: Indica se o cliente assina várias linhas telefônicas com a empresa: Sim, Não (se o cliente não estiver inscrito no serviço de telefone residencial, isso será Não).<br>
- `Serviço de Internet`: Indica se o cliente assina o serviço de Internet com a empresa: Sim, Não.<br>
- `Tipo de Internet`: Indica o tipo de conexão à Internet do cliente: DSL, Fibra Ótica, Cabo (se o cliente não estiver inscrito no serviço de Internet, isso será Nenhum).<br>
- `Custo Médio Mensal de Download em GB`: Indica o volume médio de download em gigabytes do cliente, calculado até o final do trimestre especificado acima (se o cliente não estiver inscrito no serviço de Internet, isso será 0).<br>
- `Segurança Online`: Indica se o cliente assina um serviço adicional de segurança online fornecido pela empresa: Sim, Não (se o cliente não estiver inscrito no serviço de Internet, isso será Não).<br>
- `Backup Online`: Indica se o cliente assina um serviço adicional de backup online fornecido pela empresa: Sim, Não (se o cliente não estiver inscrito no serviço de Internet, isso será Não).<br>
- `Plano de Proteção de Dispositivos`: Indica se o cliente assina um plano adicional de proteção de dispositivos para seu equipamento de internet fornecido pela empresa: Sim ou Não (se o cliente não assinar o serviço de internet será Não). <br>
- `Suporte Técnico Premium`: Indica se o cliente assina um plano adicional de suporte técnico da empresa com tempos de espera reduzidos: Sim ou Não (se o cliente não assinar o serviço de internet será Não).<br>
- `TV por Streaming`: Indica se o cliente usa seu serviço de internet para transmitir programação de televisão de um provedor terceirizado sem custo adicional: Sim ou Não (se o cliente não assinar o serviço de internet será Não).<br>
- `Filmes por Streaming`: Indica se o cliente usa seu serviço de internet para transmitir filmes de um provedor terceirizado sem custo adicional: Sim ou Não (se o cliente não assinar o serviço de internet será Não).<br>
- `Música por Streaming`: Indica se o cliente usa seu serviço de internet para transmitir música de um provedor terceirizado sem custo adicional: Sim ou Não (se o cliente não assinar o serviço de internet será Não).<br>
- `Dados Ilimitados`: Indica se o cliente pagou uma taxa mensal adicional para ter downloads/uploads ilimitados de dados: Sim ou Não (se o cliente não assinar o serviço de internet será Não).<br>
- `Contrato`: Indica o tipo de contrato atual do cliente: Mensal ou Um Ano ou Dois Anos.<br>
- `Cobrança Sem Papel`: Indica se o cliente escolheu a cobrança sem papel: Sim ou Não.<br>
- `Método de Pagamento`: "Indica como o cliente paga sua conta: Retirada Bancária ou Cartão de Crédito ou Cheque Enviado.<br>
- `Pagamento Mensal`: Indica o quanto um cliente está pagando mensalmente pelos serviços prestados pela empresa.<br>
- `Pagamento Total`: Indica as cobranças totais de um cliente, calculadas até o final do trimestre especificado.<br>
- `Total de Reembolsos`: Indica o total de reembolsos do cliente que são calculados até o final do trimestre especificado acima.<br>
- `Total de Cobranças por Dados Extras`: Indica as cobranças totais do cliente por downloads extras de dados acima dos especificados em seu plano - até o final do trimestre especificado acima.<br>
- `Total de Cobranças de Longa Distância`: Indica as cobranças totais do cliente por chamadas de longa distância acima das especificadas em seu plano - até o final do trimestre especificado acima.<br>
- `Receita Total`: Indica a receita total gerada pela empresa a partir desse cliente, calculada até o final do trimestre especificado acima. Essa métrica é calculada subtraindo o total de reembolsos e adicionando o total de cobranças extras de dados e o total de cobranças de longa distância do total de cobranças.<br>
- `Status do Cliente`: Indica o status do cliente no final do trimestre: Cancelado ou Mantido ou Novo.<br>
- `Categoria de Churn`: Uma categoria de alto nível para a razão de churn do cliente, que é perguntada quando eles saem da empresa: Atitude, Concorrente, Insatisfação, Outro, Preço (diretamente relacionado à Razão de Churn).<br>
- `Razão de Churn`: A razão específica do cliente para deixar a empresa, que é perguntada quando eles saem da empresa (diretamente relacionada à Categoria de Churn).

```python
# Libs usadas para obter os dados do banco de dados Mysql
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

# Libs que usaremos para fazer a análise descritiva
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# Carregando as variáveis de ambiente do arquivo .env
load_dotenv()

# Acessando as variáveis de ambiente carregadas
DB_USER = os.getenv('DB_USER')
DB_PASSWORD = os.getenv('DB_PASSWORD')
DB_HOST = os.getenv('DB_HOST')
DB_NAME = os.getenv('DB_NAME')

# Criaando uma conexão com o banco de dados MySQL
engine = create_engine(f'mysql+mysqlconnector://{DB_USER}:{DB_PASSWORD}@{DB_HOST}/{DB_NAME}')
```

```python
# Usando a conexão para carregar os dados em um dataframe do pandas
dataset = pd.read_sql_table('tb_customer_churn', con=engine)

# Fechando a conexão
engine.dispose()
```
